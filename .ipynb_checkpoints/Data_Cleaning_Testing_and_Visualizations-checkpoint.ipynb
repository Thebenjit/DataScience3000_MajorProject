```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, MinMaxScaler

file_path = 'OKCupid_profiles.csv'
data = pd.read_csv(file_path)

data.head()
data.info()
```

```python
# Step 2: Drop Irrelevant or Corrupted Columns
# Identify columns that are either unnamed (often created by issues during data export) or contain only numerical labels (potentially irrelevant).
irrelevant_columns = [col for col in data.columns if "Unnamed" in col or str(col).isdigit()]

# Drop the identified columns from the dataset to clean it up.
cleaned_data = data.drop(columns=irrelevant_columns)

# Display the first few rows of the cleaned data to confirm the removal of irrelevant columns.
cleaned_data.head()

# Step 3: Handle Missing Values

# Separate columns into two types:
# - Categorical columns: These are typically strings or objects, such as labels or classifications.
# - Numerical columns: These contain numerical values for statistical analysis or computations.
categorical_columns = cleaned_data.select_dtypes(include=['object']).columns
numerical_columns = cleaned_data.select_dtypes(exclude=['object']).columns

# Create imputers to fill in missing values:
# - For categorical columns, replace missing values with the most frequently occurring value in each column.
categorical_imputer = SimpleImputer(strategy='most_frequent')

# - For numerical columns, replace missing values with the median value of each column.
numerical_imputer = SimpleImputer(strategy='median')

# Apply the imputation to categorical and numerical columns:
# - Use the `fit_transform` method to compute the replacement values and apply them to the data.
cleaned_data[categorical_columns] = categorical_imputer.fit_transform(cleaned_data[categorical_columns])
cleaned_data[numerical_columns] = numerical_imputer.fit_transform(cleaned_data[numerical_columns])

# Check the distribution of values in the 'status' column to understand its content and ensure completeness.
cleaned_data['status'].value_counts()

# Map various relationship statuses to broader categories:
# - Map 'single' and 'available' to 'single'.
# - Map 'seeing someone' and 'married' to 'in a relationship'.
# - Replace 'unknown' with `None` to mark it as missing data.
relationship_mapping = {
    'single': 'single',
    'available': 'single',
    'seeing someone': 'in a relationship',
    'married': 'in a relationship',
    'unknown': None
}

# Apply the mapping to the 'status' column to standardize the values.
cleaned_data['status'] = cleaned_data['status'].map(relationship_mapping)

# Drop rows where the 'status' column still has missing values (`None`) after the mapping.
cleaned_data = cleaned_data.dropna(subset=['status'])

# Recheck the distribution of values in the 'status' column to confirm successful cleaning and mapping.
cleaned_data['status'].value_counts()

# Print the final counts of the 'status' column to confirm the changes.
print(cleaned_data['status'].value_counts())

# Print the total number of rows remaining after all cleaning steps.
print(f"Total rows after cleaning: {cleaned_data.shape[0]}")

# Display the first few rows of the fully cleaned data to verify the results.
cleaned_data.head()
```

```python
import matplotlib.pyplot as plt

# Define a function to create bar plots for categorical variables against status
def plot_categorical_vs_status(data, column, status_column='status'):
    counts = data.groupby([column, status_column]).size().unstack(fill_value=0)
    counts.plot(kind='bar', figsize=(10, 6))
    plt.title(f'{column.capitalize()} vs {status_column.capitalize()}')
    plt.xlabel(column.capitalize())
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.legend(title=status_column.capitalize())
    plt.show()

# Plot for 'diet'
plot_categorical_vs_status(cleaned_data, 'diet')

# Plot for 'drinks'
plot_categorical_vs_status(cleaned_data, 'drinks')

# Plot for 'smokes'
plot_categorical_vs_status(cleaned_data, 'smokes')

# Plot for 'drugs'
plot_categorical_vs_status(cleaned_data, 'drugs')
```

```python
import seaborn as sns

# Function to create a heatmap of relationship probabilities
def heatmap_relationship_probability(data, column, status_column='status'):
    probabilities = (
        data.groupby(column)[status_column]
        .apply(lambda x: (x == 'in a relationship').mean())
        .sort_values(ascending=False)
    )
    
    # Convert to DataFrame for heatmap plotting
    prob_df = probabilities.reset_index()
    prob_df.columns = [column, 'Probability']
    
    plt.figure(figsize=(8, 6))
    heatmap_data = prob_df.pivot_table(index=column, values='Probability')
    sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={'label': 'Probability'})
    plt.title(f'Heatmap of Relationship Probability by {column.capitalize()}')
    plt.ylabel(column.capitalize())
    plt.show()

# Heatmap for 'diet'
heatmap_relationship_probability(cleaned_data, 'diet')

# Heatmap for 'drinks'
heatmap_relationship_probability(cleaned_data, 'drinks')

# Heatmap for 'smokes'
heatmap_relationship_probability(cleaned_data, 'smokes')

# Heatmap for 'drugs'
heatmap_relationship_probability(cleaned_data, 'drugs')
```

```python
# Function to create a 2D heatmap for combinations of two factors
def two_factor_heatmap(data, factor1, factor2, status_column='status'):
    probabilities = (
        data.groupby([factor1, factor2])[status_column]
        .apply(lambda x: (x == 'in a relationship').mean())
        .unstack(fill_value=0)
    )
    
    plt.figure(figsize=(12, 8))
    sns.heatmap(probabilities, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={'label': 'Probability'})
    plt.title(f'Probability of Being in a Relationship by {factor1.capitalize()} and {factor2.capitalize()}')
    plt.xlabel(factor2.capitalize())
    plt.ylabel(factor1.capitalize())
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.show()

# Pairwise heatmaps for combinations of the factors
two_factor_heatmap(cleaned_data, 'diet', 'drinks')
two_factor_heatmap(cleaned_data, 'diet', 'smokes')
two_factor_heatmap(cleaned_data, 'diet', 'drugs')
two_factor_heatmap(cleaned_data, 'drinks', 'smokes')
two_factor_heatmap(cleaned_data, 'drinks', 'drugs')
two_factor_heatmap(cleaned_data, 'smokes', 'drugs')
```

```python
from scipy.stats import chi2_contingency

# Function to calculate and test correlation between a categorical factor and relationship status
def test_correlation_with_status(data, factor, status_column='status'):
    contingency_table = pd.crosstab(data[factor], data[status_column])
    chi2, p_value, _, _ = chi2_contingency(contingency_table)
    return {
        "Factor": factor,
        "Chi2 Statistic": chi2,
        "p-value": p_value,
        "Significant": p_value < 0.05  # Significance threshold
    }

# Test correlation for each lifestyle factor
factors = ['diet', 'drinks', 'smokes', 'drugs']
correlation_results = [test_correlation_with_status(cleaned_data, factor) for factor in factors]

# Display the results in a DataFrame for clarity
correlation_df = pd.DataFrame(correlation_results)
correlation_df
```

```python
import statsmodels.api as sm
import numpy as np

# Function to calculate proportions and confidence intervals for each category in a factor
def calculate_confidence_intervals(data, factor, status_column='status'):
    proportions = []
    confidence_intervals = []
    
    for category in data[factor].unique():
        subset = data[data[factor] == category]
        proportion = (subset[status_column] == 'in a relationship').mean()
        count = subset.shape[0]
        # Calculate confidence interval using binomial proportions
        ci_low, ci_high = sm.stats.proportion_confint(
            count=(subset[status_column] == 'in a relationship').sum(),
            nobs=count,
            alpha=0.05,  # 95% CI
            method='wilson'
        )
        proportions.append(proportion)
        confidence_intervals.append((ci_low, ci_high))
    
    return pd.DataFrame({
        'Category': data[factor].unique(),
        'Proportion': proportions,
        'CI_Low': [ci[0] for ci in confidence_intervals],
        'CI_High': [ci[1] for ci in confidence_intervals]
    })

# Adjust the error bar calculations to ensure non-negative values
def plot_confidence_intervals_fixed(data, factor, status_column='status'):
    ci_data = calculate_confidence_intervals(data, factor, status_column)
    ci_data = ci_data.sort_values(by='Proportion', ascending=False)
    
    yerr = np.abs([
        ci_data['Proportion'] - ci_data['CI_Low'],
        ci_data['CI_High'] - ci_data['Proportion']
    ])
    
    plt.figure(figsize=(12, 6))
    plt.errorbar(
        ci_data['Category'], 
        ci_data['Proportion'], 
        yerr=yerr, 
        fmt='o', capsize=5, color='blue', label='95% CI'
    )
    plt.axhline(y=ci_data['Proportion'].mean(), color='red', linestyle='--', label='Mean Proportion')
    plt.title(f'Proportion of Being in a Relationship by {factor.capitalize()} with 95% Confidence Intervals')
    plt.xlabel(factor.capitalize())
    plt.ylabel('Proportion in a Relationship')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Plot confidence intervals for diet, drinks, and drugs with fixed error bars
plot_confidence_intervals_fixed(cleaned_data, 'diet')
plot_confidence_intervals_fixed(cleaned_data, 'drinks')
plot_confidence_intervals_fixed(cleaned_data, 'drugs')
```

```python
# Importing required libraries and modules:
# - `StandardScaler`, `PolynomialFeatures`: For feature scaling and generating polynomial features.
# - `train_test_split`, `cross_val_score`: For splitting data into training and testing sets, and for cross-validation.
# - `LogisticRegression`: To build a logistic regression model.
# - `SMOTE`: To address class imbalance by generating synthetic samples.
# - `Pipeline`: To streamline the creation of machine learning pipelines.
# - `classification_report`, `roc_auc_score`, `confusion_matrix`: For evaluating model performance.
# - `RandomForestClassifier`: For building a random forest model.
# - `numpy`: For numerical operations.
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Step 1: Handle Class Imbalance with SMOTE
# Extract features (`X`) and target (`y`) from the cleaned dataset.
X = cleaned_data[['diet', 'drinks', 'smokes', 'drugs']]  # Features
y = cleaned_data['status']  # Target

# Convert the target variable `y` to numeric values:
# - 0 for 'single'
# - 1 for 'in a relationship'
y = y.map({'single': 0, 'in a relationship': 1})

# Encode categorical variables in the feature set (`X`) using one-hot encoding.
# This converts categorical variables into binary dummy variables.
X = pd.get_dummies(X, drop_first=True)

# Ensure all features in `X` are of type float, as SMOTE requires numeric inputs.
X = X.astype(float)

# Split the data into training and testing sets:
# - 70% training, 30% testing.
# - Use `stratify=y` to ensure the class distribution is preserved in both sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)

# Initialize the SMOTE instance to address class imbalance.
smote = BorderlineSMOTE(random_state=42)

# Apply SMOTE to the training set to generate synthetic samples for the minority class.
# This helps balance the class distribution.
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Step 2: Create Interaction Terms and Polynomial Features
# Initialize PolynomialFeatures to generate second-degree polynomial features.
# Set `interaction_only=False` to include all polynomial combinations.
# Set `include_bias=False` to exclude the constant bias term.
poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)

# Generate polynomial features for the training set.
X_train_poly = poly.fit_transform(X_train_resampled)

# Transform the testing set using the same polynomial feature generator.
X_test_poly = poly.transform(X_test)

# Step 3: Train Logistic Regression Model as Baseline
# Initialize a logistic regression model:
# - Set `max_iter=1000` to allow more iterations for convergence.
# - Use `class_weight='balanced'` to handle any remaining class imbalance.
logistic_model = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)

# Train the logistic regression model using the resampled training set with polynomial features.
logistic_model.fit(X_train_poly, y_train_resampled)

# Evaluate the logistic regression model on the test set.
# Predict class labels for the test set.
y_pred_logistic = logistic_model.predict(X_test_poly)

# Calculate the ROC-AUC score, which measures the ability of the model to distinguish between classes.
logistic_auc = roc_auc_score(y_test, logistic_model.predict_proba(X_test_poly)[:, 1])

# Print the evaluation metrics for the logistic regression model:
# - Classification report: Precision, recall, F1-score for each class.
# - ROC-AUC score: Overall performance metric.
# - Confusion matrix: Actual vs predicted classes.
print("Logistic Regression Results:")
print(classification_report(y_test, y_pred_logistic))
print(f"ROC-AUC Score: {logistic_auc:.4f}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred_logistic)}")

# Step 4: Train a Random Forest Model
# Initialize a random forest classifier:
# - Use 100 estimators (trees).
# - Set `class_weight='balanced'` to address class imbalance.
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')

# Train the random forest model on the resampled training set with polynomial features.
rf_model.fit(X_train_poly, y_train_resampled)

# Evaluate the random forest model on the test set.
# Predict class labels for the test set.
y_pred_rf = rf_model.predict(X_test_poly)

# Calculate the ROC-AUC score for the random forest model.
rf_auc = roc_auc_score(y_test, rf_model.predict_proba(X_test_poly)[:, 1])

# Print the evaluation metrics for the random forest model:
# - Classification report.
# - ROC-AUC score.
# - Confusion matrix.
print("\nRandom Forest Results:")
print(classification_report(y_test, y_pred_rf))
print(f"ROC-AUC Score: {rf_auc:.4f}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred_rf)}")

# Feature Importance for Random Forest
# Extract feature names from the polynomial feature generator.
feature_names = poly.get_feature_names_out(X.columns)

# Create a DataFrame to store feature importance scores from the random forest model.
feature_importances = pd.DataFrame({'Feature': feature_names, 'Importance': rf_model.feature_importances_})

# Sort the features by their importance in descending order.
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

# Display the top 10 most important features.
print(feature_importances.head(10))
```
